```
from transformers import pipeline
```

```
pipe = pipeline("text-generation", model="Alireza1044/dwight_bert_lm")
```

```
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```
tokenizer = AutoTokenizer.from_pretrained("Alireza1044/dwight_bert_lm")
model = AutoModelForCausalLM.from_pretrained("Alireza1044/dwight_bert_lm")
```

```
text = pipe(
    "You look gorgeous babe, i love",
    num_return_sequences=1,
    max_length=30,
    top_p=0.92,
    temperature=0.85
)

print(text[0]["generated_text"])
```

```
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
```

```
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
```

```
text_to_summarize = (
    """The United States is a country in North America. It is the third largest country in the world by land area and the fourth largest by population.
    The United States is a federal republic composed of 50 states, a federal district, five major unincorporated territories, and various minor islands.
    The capital is Washington, D.C., and the most populous city is New York City. The United States is a member of the United Nations, NATO, the G7, the G20,
    and the Organisation for Economic Co-operation and Development."""
)
```

```
summary = summarizer(text_to_summarize,min_length=25,max_length=55)
print(summary)
```
